# k-nn알고리즘 사용하기

```python
df['한파특보(O/X)'].value_counts()
```

```python
df['한파특보(O/X)']
cold_O = (df['한파특보(O/X)'] == '주의보') |  (df['한파특보(O/X)'] == '경보')
df['한파(O/X)'] = df['한파특보(O/X)']
df['한파(O/X)'][cold_O] = 'O'
df['한파(O/X)'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
X = df.iloc[:, 2:6]
y = df['한파(O/X)']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.3, random_state = 10)
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
y_hat = knn.predict(X_test)
```

```python
knn_report = metrics.classification_report(y_test, y_hat)
print(knn_report)
print(knn.score(X_train, y_train))
print(knn.score(X_test, y_test))
```

```python
from sklearn.model_selection import cross_val_score
k_range = range(1,25)
k_scores = []
for k in k_range :
	knn_fold = KNeighborsClassifier(n_neighbors = k)
	scores = cross_val_score(knn_fold, X, y, cv = 5, scoring = "accuracy")    # 10-fold cross-validation
	k_scores.append(scores.mean())

plt.plot(k_range, k_scores)
plt.xlabel('K의 값')
plt.ylabel('교차검증 정확도')
plt.show()

print('가장 높은 교차검증 정확도:', max(k_scores))
print('교차검증 정확도가 가장 높은 K의 값 =', k_scores.index(max(k_scores)) + 1)
```

```python
knn_21 = KNeighborsClassifier(n_neighbors = 21)
knn_21.fit(X_train, y_train)
y_hat_21 = knn_21.predict(X_test)
knn_report_21 = metrics.classification_report(y_test, y_hat_21)
print(knn_report_21)
print(knn_21.score(X_train, y_train))
print(knn_21.score(X_test, y_test))
```
